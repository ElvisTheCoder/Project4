```python
# Delete rows in which there is no value in the Year column.
df_un = df_un.dropna(subset=['Year'])
```

```python
df_un.rename(columns={"Region_subregion_country_area ": "LOCATION"}, inplace=True)
df_un
```

```python
df_un["Year"] = df_un["Year"].astype(int)
# Change Region_subregion_country_area column name to LOCATION
```

```python
# Reset all indexs for all dataframes
df_edu.reset_index(drop=True, inplace=True)
df_G20.reset_index(drop=True, inplace=True)
df_GDP.reset_index(drop=True, inplace=True)
df_Life.reset_index(drop=True, inplace=True)
df_pop.reset_index(drop=True, inplace=True)
df_un.reset_index(drop=True, inplace=True)
df_unemp.reset_index(drop=True, inplace=True)
```

```python
# Merge all dataframes into one on the LOCATION and Year columns
merge1 = pd.merge(df_G20, df_edu, on=["LOCATION", "Year"], how="left")
merge2 = pd.merge(merge1, df_GDP, on=["LOCATION", "Year"], how="left")
merge3 = pd.merge(merge2, df_Life, on=["LOCATION", "Year"], how="left")
merge4 = pd.merge(merge3, df_pop, on=["LOCATION", "Year"], how="left")
merge5 = pd.merge(merge4, df_un, on=["LOCATION", "Year"], how="left")
merge6 = pd.merge(merge5, df_unemp, on=["LOCATION", "Year"], how="left")
# Drop INDICATOR_x	SUBJECT_x	MEASURE_x	FREQUENCY_x columns
merge6 = merge6.drop(columns=["INDICATOR_x", "SUBJECT_x", "MEASURE_x", "FREQUENCY_x"])
merge6 = merge6.drop(columns=["INDICATOR_y", "SUBJECT_y", "MEASURE_y", "FREQUENCY_y"])
merge6 = merge6.drop(columns=["INDICATOR", "SUBJECT", "MEASURE", "FREQUENCY"])
merge6 = merge6.drop(columns=["Unnamed: 0_x", "Type", "Unnamed: 0_y"])
merge6 = merge6[["LOCATION", "Year", "G20 Value", "Education Value", "GDP Value", "LifeExp Value", "Population Value", "TotalFertilityRate(LiveBirthsPerWoman)", "InfantMortalityRate(InfantDeathsPer1000LiveBirths)", "Unemployment Value"]]
merge6
```

```python
merge6["Population Value"] = merge6["Population Value"] * 1000000
merge6["GDP Value"] = merge6["GDP Value"] * 1000000
```

```python
merge6["GDP_per_capita"] = merge6["GDP Value"] / merge6["Population Value"]
```

```python
# Create GDP per capita column
merge6["GDP per capita"] = merge6["GDP Value"] / merge6["Population Value"]
merge6.head(30)
```
