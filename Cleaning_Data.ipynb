```python
UN_path = "..\project 4\Data_sets\DEMOGRAPHIC_INDICATOR.csv"
GDP_path = "..\project 4\Data_sets\GDP_perCapita_USD.csv"
LifeExp_path = "..\project 4\Data_sets\Life_expectancy_at_birth.csv"
Population_path = "..\project 4\Data_sets\Population.csv"
Edu_path = "..\project 4\Data_sets\Tertiary_Edu_MenNwomen_all.csv"
Unemployment_path = "../project 4/Data_sets/Unemployment.csv"
```

```python
UN = pd.read_csv(UN_path)
UN
UN.to_csv('UN.csv', encoding='utf-8')
UN
```

```python
GDP = pd.read_csv(GDP_path)
GDP= GDP.drop(columns='Flag Codes')
GDP.head()
GDP.to_csv('GDP.csv', encoding='utf-8')
GDP
```

```python
LifeExp = pd.read_csv(LifeExp_path)
LifeExp = LifeExp.drop(columns='Flag Codes')
LifeExp.head()
LifeExp.to_csv('LifeExp.csv', encoding='utf-8')
LifeExp
```

```python
Population = pd.read_csv(Population_path)
Population = Population.drop(columns='Flag Codes')
Population.head()
Population.to_csv('Population.csv', encoding='utf-8')
Population
```

```python
Edu = pd.read_csv(Edu_path)
Edu = Edu.drop(columns='Flag Codes')
Edu.head()
Edu.to_csv('Edu.csv', encoding='utf-8')
Edu
```

```python
Unemployment = pd.read_csv(Unemployment_path)
Unemployment = Unemployment.drop(columns='Flag Codes')
Unemployment.head()
Unemployment.to_csv('Unemployment.csv', encoding='utf-8')
Unemployment
```

```python
GDP_LifeExp = pd.merge(GDP, LifeExp, how="left", on=["LOCATION", "INDICATOR","SUBJECT","MEASURE","FREQUENCY", "TIME", "Value"])
GDP_LifeExp
```

```python
Pop_Edu = pd.merge(Population, Edu, how="left", on=["LOCATION", "INDICATOR","SUBJECT","MEASURE","FREQUENCY", "TIME", "Value"])
Pop_Edu
```

```python
Pop_Edu_Unmp = pd.merge(Pop_Edu, Unemployment, how="left", on=["LOCATION", "INDICATOR","SUBJECT","MEASURE","FREQUENCY", "TIME", "Value"])
Pop_Edu_Unmp
```

```python
G20 = pd.merge(Pop_Edu_Unmp, GDP_LifeExp, how="left", on=["LOCATION", "INDICATOR","SUBJECT","MEASURE","FREQUENCY", "TIME", "Value"])
G20
G20.to_csv('G20.csv', encoding='utf-8')
G20
```
